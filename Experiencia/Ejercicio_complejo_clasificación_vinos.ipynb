```python
X = vinos.iloc[:, 0:4]
y = vinos['calidad']


from sklearn.model_selection import train_test_split


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.35, random_state = 101)

print(f'Número total de muestras en el dataset: {len(X)}')
print("*****"*10)
print(f'Muestras para entrenamiento: {len(X_train)}')
print(f'Tamaño de X_train: {X_train.shape}')
print("*****"*10)
print(f'Muestras para validación: {len(X_test)}')
print(f'Tamaño de X_test: {X_test.shape}')
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
# Una buena práctica es ajustar los datos para que tengan media cero y varianza uno. Es decir, alteramos linealmente los datos para que estén escalados entre -1 y 1 de tal forma que el entrenamiento de los modelos sea más rápido y mas estable numéricamente. Con esto se evitan problemas de explosión de gradientes.
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
from sklearn.svm import SVC

model = SVC()
model.fit(X_train, y_train)
pred = model.predict(X_test)

train_score = model.score(X_train, y_train)
print(f'Train score of trained model: {train_score*100}')

test_score = model.score(X_test, y_test)
print(f'Test score of trained model: {test_score*100}')
```

```python
# Una forma de medir el desempeño de un modelo de clasificación es mediante el uso de la matriz de confusión que ayuda a saber qué tan bien fueron clasificados los datos
# en la matriz de confusion se saben cuantos fueron verdaderos positivos y verdaderos negativos (clasificaciones correctas)
# falsos positivos y falsos negativos (clasificaciones incorrectas).
from sklearn.metrics import confusion_matrix

conf_matrix = confusion_matrix(pred, y_test)
print(f'Confussion Matrix: \n{conf_matrix}\n')
sns.heatmap(conf_matrix, annot=True)
```

```python

print("Unique values in 'calidad' column:", vinos['calidad'].unique())
calidad_mapping = {5: 0, 6: 1} # Example mapping based on your output [5, 6]
vinos_filtered = vinos[vinos['calidad'].isin(calidad_mapping.keys())].copy()

print("Shape of vinos_filtered after 'calidad' filtering:", vinos_filtered.shape)

# Create the target column on the filtered DataFrame
vinos_filtered["target"] = vinos_filtered["calidad"].map(calidad_mapping)

# Select relevant feature columns and the target column
feature_columns = ['acidez_fija', 'acidez_volatil', 'pH', 'alcohol']

# Drop rows where any of the selected feature columns have NaN values
# Now we use vinos_filtered for dropna
vinos_cleaned = vinos_filtered.dropna(subset=feature_columns + ['target']).copy()

# Check the shape of the cleaned DataFrame
print("Shape of vinos_cleaned after filtering and dropna:", vinos_cleaned.shape)

# Check for NaN values in the relevant columns of the original DataFrame (for debugging purposes)
print("\nNaN count in original 'vinos' DataFrame relevant columns:")
print(vinos[feature_columns + ['calidad']].isnull().sum())


# separando el dataset using the cleaned DataFrame
# Using the selected columns as features and the 'target' column as the label
X = vinos_cleaned[feature_columns].values
y = vinos_cleaned['target'].values

# preprocesando los datos para que tengan media cero y desviación de uno
sc = StandardScaler()
X = sc.fit_transform(X) # This should now work as X will not be empty

# particionando
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=101) # Added random_state for reproducibility

# Ajustando el modelo
from sklearn.linear_model import LogisticRegression # Import LogisticRegression
classifier = LogisticRegression(solver='lbfgs', multi_class="auto")
classifier.fit(X_train, y_train)
```

```python
# predicioendo los resultados
y_pred = classifier.predict(X_test)
probs_y = classifier.predict_proba(X_test) # mirando las probabilidades de las predicciones

# Imprimiendo los resultados de la clasificación del modelo:
probs_y = np.round(probs_y, 2)
# Ajustando el encabezado y la cadena de formato para que coincidan con las 2 clases
res = "{:<10} | {:<10} | {:<10} | {:<13}".format("y_test", "y_pred", "baja(%)", "media(%)\n")
res += "-"*65+"\n"
# Ajustando el zip para incluir solo las dos primeras columnas de probabilidad
res += "\n".join("{:<10} | {:<10} | {:<10} | {:<13}".format(x, y, a, b) for x, y, a, b in zip(y_test, y_pred, probs_y[:,0], probs_y[:,1]))
res += "\n"+"-"*65+"\n"
print(res)
```

```python
# Evaluando matriz de confusión
cm = confusion_matrix(y_test, y_pred)
print(cm)
```

```python
ax = plt.axes()
df_cm = cm
sns.heatmap(df_cm, annot=True, annot_kws={"size": 30}, fmt='d',cmap="Blues", ax = ax )
ax.set_title('Confusion Matrix')
plt.show()
```
