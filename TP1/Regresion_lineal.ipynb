```python

import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import numpy as np
import random

# === 1. Generar datos aleatorios de consumo (km/l) ===
random.seed(42)  # Para que los n√∫meros sean reproducibles
consumo_km_l = sorted([round(random.uniform(10, 20), 2) for _ in range(30)])  # 30 autos

# === 2. Generar peso basado en el consumo con algo de ruido ===
# A mayor consumo, menor peso (relaci√≥n inversa)
peso_autos = [round(2000 - c * 50 + random.uniform(-50, 50)) for c in consumo_km_l]

# Convertir consumo a matriz para usar en sklearn
X = [[c] for c in consumo_km_l]
y = peso_autos

# === 3. Crear y entrenar el modelo ===
modelo = LinearRegression()
modelo.fit(X, y)
predicciones = modelo.predict(X)

# === 4. Mostrar los coeficientes ===
print("Pendiente (coeficiente):", modelo.coef_[0])
print("Intersecci√≥n:", modelo.intercept_)

# === 5. Graficar ===
plt.scatter(consumo_km_l, peso_autos, color='green', label='Datos reales')
plt.plot(consumo_km_l, predicciones, color='red', label='L√≠nea de regresi√≥n')
plt.xlabel('Consumo (km por litro)')
plt.ylabel('Peso del auto (kg)')
plt.title('Relaci√≥n entre consumo y peso del auto (datos aleatorios)')
plt.legend()
plt.grid(True)
plt.show()

```
