```python
#importando bibliotecas (libreries)
import pandas as pd
import sklearn
import seaborn as sns
from  sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

#cargando el dataset
df_helados = pd.read_csv("IceCreamData.csv")

# Extrayendo la informacion del Dataset
X = df_helados[["Temperature"]]
y = df_helados["Revenue"]

# Particionando en dos conjuntos_train y_test
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.25)

#creando el modelo
regresor = LinearRegression(fit_intercept = True)
#Entrenarlo con los datos
regresor.fit(x_train, y_train)

#El modelo tendra dos parametros (A y B fueron llamados en el documento)
print(f"La pendiente del modelo (B) es: {regresor.coef_}, el intercepto (A) es: {regresor.intercept_}")
```

```python
y_pred = regresor.predict(x_test)
error1 = y_test - y_pred
error2 = y_pred - y_test
sns.distplot(error1)
sns.distplot(error2)
```

```python
#importando bibliotecas (libreries)
import pandas as pd
import sklearn
import seaborn as sns
from  sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

#cargando el dataset
df_helados = pd.read_excel("Regrecion_lineal_multiple.xlsx")

# Extrayendo la informacion del Dataset
X = df_helados[["x1", "x2","x3", "x4"]]
y = df_helados["y"]


# Particionando en dos conjuntos: entrenamiento (train) y prueba (test)
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.25)

#creando el modelo
regresor = LinearRegression(fit_intercept = True)
#Entrenarlo con los datos
regresor.fit(x_train, y_train)

#El modelo tendra dos parametros (A y B fueron llamados en el documento)
print(f"La pendiente del modelo (B) es: {regresor.coef_}, el intercepto (A) es: {regresor.intercept_}")
```

```python
y_pred = regresor.predict(x_test)
error1 = y_test - y_pred
error2 = y_pred - y_test
sns.distplot(error1)
sns.distplot(error2)
```
