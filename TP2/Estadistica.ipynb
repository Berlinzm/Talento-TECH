```python
import pandas as pd

# create a dataset
# Let´s build the values lists
value_n_list = ["Deiby",
                "Daniel",
                "Daira",
                "Carolina",
                ]
print(value_n_list)

value_w_list = [57, 64, 60, 58,] # peso en Kg

value_h_list = [1.45, 1.77, 1.55, 1.62,] # altura en metros


data_dict = {"Name": value_n_list, "Weight": value_w_list, "Height": value_h_list,}
             #"BMI": value_bmi_list}

print(data_dict)
df = pd.DataFrame(data_dict)
print(df)
df
```
